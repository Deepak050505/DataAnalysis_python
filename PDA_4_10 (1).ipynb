# One-Hot-Encoding:

Encoding

Converting categorical data into numerical data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\Salary_EDA.csv")
df.head()
```

Filter Categorical feature

```python
categorical_cols=['Education Level']
```

Define and apply Encoder

```python
encoder=OneHotEncoder(drop=None,sparse_output=False)
```

```python
encoder_data=encoder.fit_transform(df[categorical_cols])
```

```python
print(encoder_data)
```

```python
encoded_df=pd.DataFrame(encoder_data, columns=encoder.get_feature_names_out(categorical_cols))
encoded_df.head()
```

```python
encoded_df.drop(columns=['Education Level_nan'],inplace=True)
encoded_df.head()
```

```python
final_df=pd.concat([df,encoded_df],axis=1)
final_df.head()
```
